# İŞLEMLER

* Eleman Temelli

Bu işlemler tensorün er elemanınaayrı ayrı uygulanır.

```
def naive_add(x, y):
  assert len(x.shape) == 2#x ve y 2 byutlu numpy array

  x = x.copy()

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[i, j]
  return x    
```

```
x = np.array([[7, 14, 21],
             [7, 14, 21],
             [7, 14, 21]])

y = np.array([[1, 1, 1],
              [2, 2 ,2],
              [3, 3, 3]])

z = naive_add(x, y)
z
```

```
def naive_relu(x):
  assert len(x.shape) == 2#x ve y 2 byutlu numpy array

  x = x.copy()

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] =max(x[i, j],0)
  return x    
```

```
# Yukarıdaki işleminnaynısını numpy kullanarak yaptık.
import numpy as np
x = np.array([[7, 14, 21],
              [7, 14, 21],
              [7, 14, 21]])

y = np.array([[1, 1, -1],
              [-2, -2 ,-2],
              [3, 3, -3]])
z = naive_relu(y)
z
```

* Yayma Operasyonu



```
def naive_add_matrix_and_vector(x, y):
  assert len(x.shape) == 2
  assert len(y.shape) == 1
  assert x.shape[1] == y.shape[0]

  x = x.copy()

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[j]
  return x
```

```
x = np.array([[7, 14, 21],
              [7, 14, 21],
              [7, 14, 21]])

y = np.array([1, 2, 3])

z = naive_add_matrix_and_vector(x, y)
z
```
